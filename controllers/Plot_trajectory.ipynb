```python
import numpy as np
import os
from matplotlib import pyplot as plt
```

```python
data = {}
colors = ["", "red", "blue", "black", "yellow", "purple"]
```

```python
plt.figure(figsize=(16, 12))
base1 = "./heterogenous_robot/output/"
base2 = "./goal_controller/output/"
for file in os.listdir(base1):
    index = int(file[0])
    data = np.load(base1 + file)
    x,y = data.T
    if file[2] == "0":
        plt.plot(x, -y, color = colors[index], label = "robot_" + str(index))
    else:
        plt.plot(x, -y, color = colors[index])
for file in os.listdir(base2):
    name = file.split("_")
    data = np.load(base2 + file)
    x,y = data.T
    if name[2][0] == "0":
        plt.plot(x, -y, color = "green", label = "Goal_" + name[1])
    else:
        plt.plot(x, -y, color = "green")
# plt.axis('equal')
plt.xlabel("X coordinate", fontsize=24)
plt.ylabel("Y coordinate", fontsize=24)
plt.title("XY Trajectory of robots", fontsize=24)
plt.legend()
plt.show()
```

```python
for file in os.listdir(base1):
    index = int(file[0])
    data = np.load(base1 + file)
#     print(data)
    x,y = data.T
#     print(x,y)
    distance = 0
    for i in range(len(x)):
        if i + 1 < len(x):
            xdist = abs(x[i+1] - x[i])
            ydist = abs(y[i+1] - y[i])
            distance += np.sqrt(xdist**2 + ydist**2)
    plt.scatter(index, distance)
plt.xlabel("Epuck #", fontsize=24)
plt.ylabel("Total path distance", fontsize=24)
plt.title("Epuck distances", fontsize=24)
plt.legend()
plt.show()
```
