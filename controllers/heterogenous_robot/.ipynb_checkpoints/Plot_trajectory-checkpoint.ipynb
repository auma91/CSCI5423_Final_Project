```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
data = {}
colors = ["", "red", "blue", "black", "yellow", "green"]
for i in range(1, 6):
    data[i] = np.load(str(i) + ".npy")
data[1][:10]
```

```python
x,y = data[1].T
y[:10]
```

```python
plt.figure(figsize=(16, 12))
for i in range(1, 6):
    x,y = data[i].T
    plt.plot(x, y, color= colors[i], label ="robot_" + str(i))
plt.axis('equal')
plt.legend()
plt.show()
```
